```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[4,10,6])
#this will add a heading in chart
plt.title("my data")
#this will add x and y label name in chart
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
#plotting two plot in one chart
from matplotlib import style
x1=[1,2,3]
y1=[4,5,9]
x2=[2,5,7]
y2=[10,9,2]
style.use('ggplot')
plt.plot(x1,y1,label="first chart")
plt.plot(x2,y2,label="second chart")
plt.legend()
plt.show()
```
